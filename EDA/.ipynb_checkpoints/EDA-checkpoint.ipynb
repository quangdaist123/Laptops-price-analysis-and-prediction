```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
from math import ceil
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
pd.options.display.max_rows
```

```python
df = pd.read_csv('tidydata_filtered_missing.csv')
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

# Tỉ lệ missing

```python
# Tỉ lệ missing
labels = df.columns
na_values = df.isna().sum().values
not_na_values = df.shape[0] - df.isna().sum().values

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(figsize=(15,10))

ax.bar(labels, na_values, width, label='NA')
ax.bar(labels, not_na_values, width,  bottom=na_values,
       label='Not NA')

ax.set_ylabel('Số lượng')
ax.set_title('Số lượng NA và Not NA')
ax.legend()


_ = plt.xticks(rotation=75)
plt.savefig('missing.png')
plt.show()
```

```python
(100*df.isna().sum() / df.shape[0]).sort_values(ascending=False)#.to_csv('missing.csv')
```

# Miền giá trị

```python
quanti_cols = ['weight', 'thickness', 'width', 'length', 'used_warranty', 'used_price', 'cpu_speed', 'max_cpu_speed']
```

```python
quali_df = df.drop(columns=quanti_cols)
```

```python
# Biến rời rạc
for column in quali_df:
    _ = sns.countplot(x=column, data=df)
    __ = _.set_xticklabels(_.get_xticklabels(), rotation=60)
    plt.show()
```

```python
# Biến liên tục
for column in quanti_cols:
    if column == 'used_price':
        continue
        
    _ = sns.distplot(x=df[column])
    __ = plt.xlabel(column)
#     labels = [str(int(item/1e6)) + ' tr' for item in _.get_yticks()]
#     __ = _.set_yticklabels(labels)
    plt.show()
```

# Tương quan

```python
# Biến rời rạc
f, axes = plt.subplots(6, 5, figsize=(30,30))
f.subplots_adjust(wspace=0.5)
axes = axes.flatten()

for i, column in enumerate(quali_df.columns):
    
    _ = sns.boxplot(x=column, y='used_price', data=df, ax=axes[i-1])
    __ = _.set_xticklabels(_.get_xticklabels(), rotation=60)

    labels = [str(int(item/1e6)) + ' tr' for item in _.get_yticks()]
    __ = _.set_yticklabels(labels)

f.tight_layout()
plt.savefig('categorical.png')
```

```python
# Biến rời rạc

for column in quali_df.columns:
    # Vẽ boxplot với x là các biến rời rạc, y là 'used_price'
    _ = sns.boxplot(x=column, y='used_price', data=df)
    __ = _.set_xticklabels(_.get_xticklabels(), rotation=60)
    
    # Custom nhãn của yticks
    labels = [str(int(item/1e6)) + ' tr' for item in _.get_yticks()]
    __ = _.set_yticklabels(labels)
    plt.show()
```

```python
# Biến liên tục
for column in quanti_cols:
    if column == 'used_price':
        continue
        
    _ = sns.regplot(x=column, y='used_price', data=df)
    labels = [str(int(item/1e6)) + ' tr' for item in _.get_yticks()]
    __ = _.set_yticklabels(labels)
    plt.show()
```
