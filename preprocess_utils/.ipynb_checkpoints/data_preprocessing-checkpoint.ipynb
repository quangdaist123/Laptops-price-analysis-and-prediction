```python
import re
def preprocess_RAM_tiki(string_in):
    if not string_in or str(string_in) == "nan":
        return np.nan
    else:
        # Lấy dung lượng RAM
        string_in = string_in.strip()
        
        if string_in[:2].isdigit():
            ram_capacity = string_in[:2]
        else:
            ram_capacity = string_in[0]
    
        # Lấy loại RAM
        types_of_RAM = ["LPDDR4X", "DDR4"]
        ram_type = np.nan
        for type in types_of_RAM:
            if type in string_in:
                ram_type = type
                break
        
        # Lấy bus RAM
        string_in = string_in.lower()
        bus = re.match(r'\d+\ ?mhz', string_in)
        if bus:
            bus = int(bus.replace('mhz', ''))
        else:
            bus = np.nan
        
    print(ram_capacity, ram_type, bus)
```
