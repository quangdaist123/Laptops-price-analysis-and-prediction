```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
from math import ceil
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
pd.options.display.max_rows
```

```python
df = pd.read_csv('data.csv')
```

# Tỉ lệ missing

```python
# Tỉ lệ missing
labels = df.columns
na_values = df.isna().sum().values
not_na_values = df.shape[0] - df.isna().sum().values

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(figsize=(15,10))

ax.bar(labels, na_values, width, label='NA')
ax.bar(labels, not_na_values, width,  bottom=na_values,
       label='Not NA')

ax.set_ylabel('Số lượng')
ax.set_title('Số lượng NA và Not NA')
ax.legend()

#plt.figure(figsize=(15,10))
_ = plt.xticks(rotation=75)
plt.show()
```

```python
objectColumns = [column for column in df.columns if (df.dtypes[column] == 'O')]
```

```python
df['CPU'].value_counts().plot(kind='bar', x=df['CPU'].value_counts().index)
plt.xlabel('CPU')
plt.ylabel('Số lượng')
```

```python
df['Loại RAM'].value_counts().plot(kind='bar', x=df['Loại RAM'].value_counts().index)
plt.xlabel('Loại RAM')
plt.ylabel('Số lượng')
```

# Tương quan

```python
for column in objectColumns:
    sns.boxplot(x=column, y='Giá máy mới', data=df)
    plt.show()
```

```python
categorical_features = df.select_dtypes(exclude="number").columns

fig, axes = plt.subplots(nrows=ceil(len(categorical_features) / 2), ncols=2, figsize=(40, 75))
fig.subplots_adjust(wspace=1)
for i in range(len(categorical_features)):
    grouped = df.groupby(categorical_features[i])
    keys = list(grouped.groups.keys())
    anova_result = f_oneway(*[grouped.get_group(key)["Giá máy mới"] for key in keys])
    plot = sns.boxplot(ax=axes[i // 2][i % 2], x=categorical_features[i], y="Giá máy mới", data=df)
    plot.set(title=anova_result)
plt.show()
```
